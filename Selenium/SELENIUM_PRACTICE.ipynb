# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.shine.com/')
```

```python
job_title=driver.find_element(By.CLASS_NAME,"form-control  ")
job_title.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Banglore')
```

```python
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
title_tags=driver.find_elements(By.XPATH,'//strong[@class="jobCard_pReplaceH2__xWmHg"]')
for i in title_tags:
    title=i.text
    job_title.append(title)


location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags:
    location=i.text
    job_location.append(location)

company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in company_tags:
    company=i.text
    company_name.append(company)


experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags:
    experience=i.text
    experience_required.append(experience)

```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company':company_name,'Experience':experience_required})
df
```

# Q2: Write a python program to scrape data for ““Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver1=webdriver.Chrome()
```

```python
driver1.get("https://www.naukri.com/")
```

```python
job_title1=driver1.find_element(By.CLASS_NAME,"suggestor-input ")
job_title1.send_keys('Data Analyst')
```

```python
location1=driver1.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location1.send_keys('Banglore')
```

```python
search1=driver1.find_element(By.CLASS_NAME,"qsbSubmit")
search1.click()
```

```python
job_title1=[]
job_location1=[]
company_name1=[]
experience_required1=[]
```

```python
title_tags1=driver1.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags1:
    title1=i.text
    job_title1.append(title1)


location_tags1=driver1.find_elements(By.XPATH,'//div[@class="job-details "]/span[3]')
for i in location_tags1:
    location1=i.text
    job_location1.append(location1)

company_tags1=driver1.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[2]/span/a[1]')
for i in company_tags1:
    company1=i.text
    company_name1.append(company1)


experience_tags1=driver1.find_elements(By.XPATH,'//div[@class="job-details "]/span[1] /span/span')
for i in experience_tags1:
    experience1=i.text
    experience_required1.append(experience1)
```

```python
print(len(job_title1),len(job_location1),len(company_name1),len(experience_required1))
```

```python
df1=pd.DataFrame({'Title1':job_title1,'Location1':job_location1,'Company_Name1':company_name1,'Experience1':experience_required1})
df1
```
