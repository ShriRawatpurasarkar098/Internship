# Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter.
You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results.
You have to scrape the job-title, job-location, company name, experience required..

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver1=webdriver.Chrome()
```

```python
driver1.get("https://www.naukri.com/")
```

```python
job_title1=driver1.find_element(By.CLASS_NAME,"suggestor-input ")
job_title1.send_keys('Data Scientist')
```

```python
location1=driver1.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location1.send_keys('Delhi/NCR')
```

```python
search1=driver1.find_element(By.CLASS_NAME,"qsbSubmit")
search1.click()
```

```python
salaryfilter=driver1.find_element(By.XPATH,'/html/body/div/div/main/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]/label/p/span[1]')
salaryfilter.click()
```

```python
job_title1=[]
job_location1=[]
company_name1=[]
experience_required1=[]
```

```python
title_tags1=driver1.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags1:
    title1=i.text
    job_title1.append(title1)


location_tags1=driver1.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[3]/div/span[3]/span')
for i in location_tags1:
    location1=i.text
    job_location1.append(location1)

company_tags1=driver1.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[2]/span/a[1]')
for i in company_tags1:
    company1=i.text
    company_name1.append(company1)


experience_tags1=driver1.find_elements(By.XPATH,'//div[@class="job-details "]/span[1] /span/span')
for i in experience_tags1:
    experience1=i.text
    experience_required1.append(experience1)
```

```python
print(len(job_title1),len(job_location1),len(company_name1),len(experience_required1))
```

```python
df1=pd.DataFrame({'Title1':job_title1,'Location1':job_location1,'Company_Name1':company_name1,'Experience1':experience_required1})
df1
```

```python
driver2.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')
```

```python
Rating=[]
Review=[]
Full_Review=[]
```

```python
start=0
end=11
for page in range(start,end):
    Rating_Tags=driver2.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in Rating_Tags:
        Rating.append(i.text)

for page in range(start,end):
    Review_Tags=driver2.find_elements(By.XPATH,'//div[@class="col EPCmJX Ma1fCG"]/div[1]/p')
    for i in Review_Tags:
        Review.append(i.text)

for page in range(start,end):
    Full_Review_Tags=driver2.find_elements(By.XPATH,'//div[@class="ZmyHeo"]')
    for i in Full_Review_Tags:
        Full_Review.append(i.text)
```

# Q6: Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quote
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver5=webdriver.Chrome()
```

```python
driver5.get('https://www.azquotes.com/')
```

```python
Quote=[]
Author=[]
Type=[]
```

```python
start=0
end=11
for page in range(start,end):
    quote_Tags=driver5.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
    for i in quote_Tags:
        Quote.append(i.text)

for page in range(start,end):
    author_Tags=driver5.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_Tags:
        Author.append(i.text)
for page in range(start,end):
    type_Tags=driver5.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_Tags:
        Type.append(i.text)
```

```python
print(len(Quote),len(Author),len(Type))
```

```python
df6=pd.DataFrame({'Quote':Quote,'Author':Author,'Type of Quote':Type})
df6
```
